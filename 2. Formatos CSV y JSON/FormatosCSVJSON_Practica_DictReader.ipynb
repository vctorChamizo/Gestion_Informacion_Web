Considerar el archivo __Agenda.csv__ que contiene una relación de actividades de carácter cultural y de ocio, que se van a celebrar en los próximos 100 días y que se están celebrando en centros Municipales.

* Tercera pregunta. Audiencia. Debe listar el conjunto de tipos de audiencia __[Campo Audiencia]__, y el usuario deberá elegir uno de ellos. Se puede asignar un número o letra a cada tipo de audiencia de manera que el usuario introduzca únicamente el número o letra seleccionado. Observar que para mostrar el tipo de audiencia, hay que tratar el valor que aparece en el campo, de manera que se elimine las apariciones de "/usuario/". Así mismo, hay casos en los que no existe una audiencia. En esos casos, cuando se realice una búsqueda, no será un campo de restricción, es decir que podrá aparecer como respuesta en las consultas si cumple los dos primeros criterios.

Se puede asignar un número o letra a cada tipo de audiencia de manera que el usuario introduzca únicamente el número o letra seleccionado. Estos intervalos no se obtienen del documento.

Como resultado el programa deberá mostrar por pantalla información de las actividades que cumplen las respuestas introducidas. Concretamente:

  6. Descripción.

```python

import csv

csvarchivo = open('Agenda.csv',encoding="utf8",errors='ignore')
entrada = csv.DictReader(csvarchivo, delimiter=";")
listaDistrito = []
listaTipo = []
listaAudiencia = []
listaResultados = []

def entradaValida(entrada,limInf,limSup):
    correcto = False;
    while correcto == False:
        if (entrada.isdigit() == True):
            if (int(entrada) >= limInf and int(entrada) <= limSup):
                correcto = True
            else:
                entrada = input("Introduce un número válido:")
        else:
            entrada = input("Introduce un número:")
    return int(entrada)


for i in entrada:
    if i['DISTRITO'] not in listaDistrito and len(i['DISTRITO']) > 0:
        listaDistrito.append(i['DISTRITO'])
    if i['TIPO'] not in listaTipo and len(i['TIPO']) > 0:     
        listaTipo.append(i['TIPO'])
    if i['AUDIENCIA'] not in listaAudiencia and len(i['AUDIENCIA']) > 0:
        listaAudiencia.append(i['AUDIENCIA'])
        

listaTipoAux = []
listaAudienciaAux = []

listaTipoAux = listaTipo.copy()
listaAudienciaAux = listaAudiencia.copy()

for i in range (len(listaTipo)):
    listaTipoAux[i] = listaTipoAux[i].replace("/contenido/actividades/", "")
    
for i in range (len(listaAudienciaAux)):
    listaAudienciaAux[i] = listaAudienciaAux[i].replace("/usuario/", "")
        
        
for i in range (len(listaDistrito)):
    print(i + 1, listaDistrito[i])

distritoElegido = input("Introduce el numero de distrito que quieres elegir: ")
distritoElegido = entradaValida(distritoElegido, 1, len(listaDistrito))

listaResultados.append(listaDistrito[distritoElegido - 1])

for i in range (len(listaTipoAux)):
    print(i + 1, listaTipoAux[i])

tipoElegido = input("Introduce el numero de tipo que quieres elegir: ")
tipoElegido = entradaValida(tipoElegido, 1, len(listaTipo))

listaResultados.append(listaTipo[tipoElegido - 1])

for i in range (len(listaAudienciaAux)):
    print(i + 1, listaAudienciaAux[i])

audienciaElegida = input("Introduce el numero de audiencia que quieres elegir: ")
audienciaElegida = entradaValida(audienciaElegida, 1, len(listaAudiencia))

listaResultados.append(listaAudiencia[audienciaElegida - 1])

for i in range (23):
    if i < 10:
        print(i + 1 , ' - ','0', i,":00-", i + 1,":00",sep="")
    else:
        print(i + 1 , ' - ', i,":00-", i + 1,":00",sep ="")
        
print(24, "- 23:00-23:59")

horaElegida = input("Introduce la franja horaria que quieres elegir: ")
horaElegida = entradaValida(horaElegida, 1, 24)

horaFinal = str(horaElegida) + ":00"

listaResultados.append(horaFinal)

from collections import defaultdict

csvarchivo.seek(0)
csvsalida = open('salida.csv','w',encoding="utf8")
salida = csv.writer(csvsalida, delimiter=";")
encontrado = False
columns = [["TITULO", "FECHA","FECHA-FIN" ,"DIAS-SEMANA", "DIAS-EXCLUIDOS", "HORA", "DESCRIPCION", "NOMBRE-INSTALACION", "TIPO", "AUDIENCIA"]]
salida.writerows(columns)
for row in entrada:
     if row['DISTRITO'] == listaResultados[0] and row['TIPO'] == listaResultados[1] and row['AUDIENCIA'] == listaResultados[2] and row['HORA'] == listaResultados[3]:
        encontrado = True
        salida.writerow((row['TITULO'], row['FECHA'], row['FECHA-FIN'], row['DIAS-SEMANA'], row['DIAS-EXCLUIDOS'], row['HORA'], row['DESCRIPCION'], row['NOMBRE-INSTALACION'], row['TIPO'], row['AUDIENCIA']))
        print("Título: ", row['TITULO'])
        print("Fecha: ", row['FECHA'])
        print("Fecha de finalización: ", row['FECHA-FIN'])
        print("Días de la semana: ", row['DIAS-SEMANA'])
        print("Días excluidos: ", row['DIAS-EXCLUIDOS'])
        print("Hora: ", row['HORA'])
        print("Descripción: ", row['DESCRIPCION'])
        print("Nombre de instalación: ", row['NOMBRE-INSTALACION'])
        row['TIPO'] = row['TIPO'].replace("/contenido/actividades/","")
        print("Tipo: ", row['TIPO'])
        row['AUDIENCIA'] = row['AUDIENCIA'].replace("/usuario/","")
        print("Audiencia: ", row['AUDIENCIA'])
        print("      ")
csvarchivo.close()
csvsalida.close()      
if encontrado == False:
    print("No existe ningún evento con las condiciones indicadas" )
else:
    print("CORRECTO!")


```

```python
import csv
import math
csvarchivo = open('Agenda.csv',encoding="utf8",errors='ignore')
entrada = csv.DictReader(csvarchivo, delimiter=";")
listaDistrito = []
listaTipo = []
listaAudiencia = []
listaResultados = []

def entradaValida(entrada,limInf,limSup):
    correcto = False;
    while correcto == False:
        if (entrada.isdigit() == True):
            if (int(entrada) >= limInf and int(entrada) <= limSup):
                correcto = True
            else:
                entrada = input("Introduce un número válido:")
        else:
            entrada = input("Introduce un número:")
    return int(entrada)

def haversine(lat1, lon1, lat2, lon2):
    rad=math.pi/180
    dlat=lat2-lat1
    dlon=lon2-lon1
    R=6372.795477598
    a=(math.sin(rad*dlat/2))**2 + math.cos(rad*lat1)*math.cos(rad*lat2)*(math.sin(rad*dlon/2))**2
    distancia=2*R*math.asin(math.sqrt(a))
    return distancia

for i in entrada:
    if i['DISTRITO'] not in listaDistrito and len(i['DISTRITO']) > 0:
        listaDistrito.append(i['DISTRITO'])
    if i['TIPO'] not in listaTipo and len(i['TIPO']) > 0:     
        listaTipo.append(i['TIPO'])
    if i['AUDIENCIA'] not in listaAudiencia and len(i['AUDIENCIA']) > 0:
        listaAudiencia.append(i['AUDIENCIA'])
        
listaTipoAux = []
listaAudienciaAux = []

listaTipoAux = listaTipo.copy()
listaAudienciaAux = listaAudiencia.copy()

for i in range (len(listaTipo)):
    listaTipoAux[i] = listaTipoAux[i].replace("/contenido/actividades/", "")
    
for i in range (len(listaAudienciaAux)):
    listaAudienciaAux[i] = listaAudienciaAux[i].replace("/usuario/", "")
        
        
for i in range (len(listaDistrito)):
    print(i + 1, listaDistrito[i])

distritoElegido = input("Introduce el numero de distrito que quieres elegir: ")
distritoElegido = entradaValida(distritoElegido, 1, len(listaDistrito))

listaResultados.append(listaDistrito[distritoElegido - 1])

for i in range (len(listaTipoAux)):
    print(i + 1, listaTipoAux[i])

tipoElegido = input("Introduce el numero de tipo que quieres elegir: ")
tipoElegido = entradaValida(tipoElegido, 1, len(listaTipo))

listaResultados.append(listaTipo[tipoElegido - 1])

for i in range (len(listaAudienciaAux)):
    print(i + 1, listaAudienciaAux[i])

audienciaElegida = input("Introduce el numero de audiencia que quieres elegir: ")
audienciaElegida = entradaValida(audienciaElegida, 1, len(listaAudiencia))

listaResultados.append(listaAudiencia[audienciaElegida - 1])

for i in range (23):
    if i < 10:
        print(i + 1 , ' - ','0', i,":00-", i + 1,":00",sep="")
    else:
        print(i + 1 , ' - ', i,":00-", i + 1,":00",sep ="")
        
print(24, "- 23:00-23:59")

horaElegida = input("Introduce la franja horaria que quieres elegir: ")
horaElegida = entradaValida(horaElegida, 1, 24)

horaFinal = str(horaElegida) + ":00"

listaResultados.append(horaFinal)

longitud = float(input("Introduce una longitud:"))
latitud = float(input("Introduce una latitud:"))
distancia = float(input("Introduce una distancia:"))

from collections import defaultdict

csvarchivo.seek(0)
csvsalida = open('salida.csv','w',encoding="utf8")
salida = csv.writer(csvsalida, delimiter=";")
encontrado = False
distanciaCalculada = 0
columns = [["TITULO", "FECHA","FECHA-FIN" ,"DIAS-SEMANA", "DIAS-EXCLUIDOS", "HORA", "DESCRIPCION", "NOMBRE-INSTALACION", "TIPO", "AUDIENCIA"]]
salida.writerows(columns)
for row in entrada:
    if(len(row['LATITUD']) == 0 or len(row['LONGITUD']) == 0 or row['LATITUD'] == 'LATITUD' or row['LONGITUD'] == 'LONGITUD'):
        distanciaCalculada = 0
    else:
        distanciaCalculada = haversine(latitud, longitud, float(row['LATITUD']), float(row['LONGITUD']))
    if row['DISTRITO'] == listaResultados[0] and row['TIPO'] == listaResultados[1] and row['AUDIENCIA'] == listaResultados[2] and row['HORA'] == listaResultados[3] and distanciaCalculada <= distancia:
        encontrado = True
        salida.writerow((row['TITULO'], row['FECHA'], row['FECHA-FIN'], row['DIAS-SEMANA'], row['DIAS-EXCLUIDOS'], row['HORA'], row['DESCRIPCION'], row['NOMBRE-INSTALACION'], row['TIPO'], row['AUDIENCIA']))
        print("Título: ", row['TITULO'])
        print("Fecha: ", row['FECHA'])
        print("Fecha de finalización: ", row['FECHA-FIN'])
        print("Días de la semana: ", row['DIAS-SEMANA'])
        print("Días excluidos: ", row['DIAS-EXCLUIDOS'])
        print("Hora: ", row['HORA'])
        print("Descripción: ", row['DESCRIPCION'])
        print("Nombre de instalación: ", row['NOMBRE-INSTALACION'])
        row['TIPO'] = row['TIPO'].replace("/contenido/actividades/","")
        print("Tipo: ", row['TIPO'])
        row['AUDIENCIA'] = row['AUDIENCIA'].replace("/usuario/","")
        print("Audiencia: ", row['AUDIENCIA'])
        print("Distancia calculada: ", distanciaCalculada)
        print("      ")
csvarchivo.close()           
csvsalida.close()
            
if encontrado == False:
    print("No existe ningún evento con las condiciones indicadas" )
else:
    print("CORRECTO!")
```
